```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
print(data["DESCR"])
```

```python
data['data']
```

```python
print(data['target_names'])
```

```python
x = pd.DataFrame(data=data['data'],columns=data['feature_names'])
```

```python
x
```

```python
data['target_names']
```

```python
y = pd.Series(data=data['target'])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=98,test_size=0.2)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt,node_ids=True,filled=True,proportion=True)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
```

### pre pruning is generally used beacause it helps the cost and time to calculate the tree. Post pruning is used for small datasets i.e we could visualize at what depth it is no longer necesaary to calculate further 


```python
param={
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
tree = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
gd = GridSearchCV(tree,param,n_jobs=-1,verbose=2,scoring="accuracy")
```

```python
gd.fit(x_train,y_train)
```

```python
gd.best_estimator_
```

```python
gd.best_params_
```

```python
y_gpred=gd.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(classification_report(y_gpred,y_test))
print(confusion_matrix(y_gpred,y_test))
print(accuracy_score(y_gpred,y_test))
```

### Diabetes Predictions

```python
from sklearn.datasets import load_diabetes
da = load_diabetes()
```

```python
print(da['DESCR'])
```

```python
print(da['feature_names'])
print(da['target'])
```

```python
x = pd.DataFrame(da['data'],columns=da['feature_names'])
y = da['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=98,test_size=0.2)
```

```python
print(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dec = DecisionTreeRegressor()
dec.fit(x_train,y_train)
```

```python
y_rpred=dec.predict(x_test)
```

```python
y_rpred
```

```python
paramr = {'criterion' : ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
          'splitter' : ['best', 'random'],
           'max_depth':[1,2,3,4,5,10,15,20,25],
            'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_test,y_rpred,))
print(mean_absolute_error(y_test,y_rpred))
print(mean_squared_error(y_test,y_rpred))
```

```python
gdr = GridSearchCV(dec,paramr,n_jobs=-1,cv = 5,scoring='neg_mean_squared_error',verbose=2)
```

```python
gdr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_test,gdr.predict(x_test)))
print(mean_absolute_error(y_test,gdr.predict(x_test)))
print(mean_squared_error(y_test,gdr.predict(x_test)))
```
