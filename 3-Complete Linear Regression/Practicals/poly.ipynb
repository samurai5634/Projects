```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = 6 * np.random.rand(100, 1) - 3
y =0.5 * X**2 + 1.5*X + 2 + np.random.randn(100, 1)
```

```python
y
X
```

```python
plt.scatter(X,y,color = 'r')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=42,train_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred  = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,r2_score
score = r2_score(y_test,lr.predict(x_test))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(score)
```

```python
sns.regplot(x=X,y=y)
```

```python
plt.plot(x_train,lr.predict(x_train),color = 'r')
plt.scatter(x_train,y_train)
plt.xlabel('x dataset')
plt.ylabel('Y')
```
