```python


n = int(input("Enter how many prime numbers you want "))
def is_prime(n):
    for i in range(2,int(n**0.5) + 1):
        if(n % i == 0):
            return False
    return 

def prime_numbers(n):
    prime = []
    num = 2
    while(len(prime) < n):
        if is_prime(num):
            prime.append(num)
        num += 1
    print(prime)

prime_numbers(n) 
```
