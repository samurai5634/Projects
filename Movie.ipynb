```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
data_1['tags'] = data_1['tags'].fillna('')
```

```python
cv.fit_transform(data_1['tags'])
```

```python
vec = cv.fit_transform(data_1['tags']).toarray()
```

```python
sin = cosine_similarity(vec)
```

```python
dis = sorted(list(enumerate(sin[25])), reverse = True,key = lambda x:x[1])
```

```python
for i in dis[0:5]:
  print (data_1.iloc[i[0]].title)
```

```python
def recommend(movies):
  index = data_1[data_1['title'] == movies].index[0]
  dist = sorted(list(enumerate(sin[index])), reverse = True,key = lambda vec:vec[1])
  for i  in dist[0:5]:
    print (data_1.iloc[i[0]].title)
```

```python
recommend('The Pursuit of Happyness')
```
