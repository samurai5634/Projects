```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,n_classes=2,n_redundant=0)
```

```python
X.shape
```

```python
X[0]
```

```python
X = pd.DataFrame(X)
```

```python
X
```

```python
Y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1], hue = Y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=13,stratify= None)
```

```python
x_train
y_train
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='poly')
svc.fit(x_train,y_train)
```

```python
PRE = svc.predict(x_test)
```

```python
PRE
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_pred=PRE,y_true=y_test))
print(classification_report(PRE,y_test))
print(confusion_matrix(PRE,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
params = { 'C' : [1,0.1,100], 'gamma': [1,0.1,100], 'kernel': ['linear'] }
```

```python
gc = GridSearchCV(estimator=svc,param_grid=params,n_jobs=-1,scoring=['accuracy'],cv = 5,refit=callable,verbose=3)
```

```python
gc
```

```python
gc.fit(x_train,y_train)
```

```python
gc.best_estimator_
```

```python
gc.best_params_
```

```python
yp = gc.predict(x_test)
```

```python
print(accuracy_score(y_pred=yp,y_true=y_test))
print(classification_report(yp,y_test))
print(confusion_matrix(yp,y_test))
```
