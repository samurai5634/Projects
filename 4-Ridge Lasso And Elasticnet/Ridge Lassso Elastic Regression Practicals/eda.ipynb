## TRAINING OF THE MODEL 


```python
df.head()
```

```python
x = df.drop(['FWI1'],axis = 1)
y = df['FWI1']
```

```python
x
```

```python
x.drop(['month'],axis=1,inplace = True)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=43,test_size=0.25)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
plt.figure(figsize=(12,8))
c = x_train.corr()
sns.heatmap(data=c,square=True,annot=True,fmt='.2f')
# plt.tight_layout()
```

```python
def feat(data,thresh):
    corr = set()
    corrmat = data.corr()
    for i in range(0,len(corrmat.columns)):
        for j in range(i):
            if(abs(corrmat.iloc[i,j]) > thresh):
                corr.add(corrmat.columns[i])
    return corr
```

```python
c = feat(x_train,0.85)
```

```python
x_train.drop(c,axis=1,inplace=True)
x_test.drop(c,axis=1,inplace=True)
```

```python
x_train.shape,x_test.shape
```

### Standardization


```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x_trainscale = std.fit_transform(x_train)
x_testscale = std.transform(x_test)
```

```python
x_trainscale
```

```python
sns.boxplot(x_trainscale)
```

```python
ax = plt.subplots(figsize = (15,5))
plt.subplot(1,2,1)
plt.xticks(rotation= 90)
sns.boxplot(x_train)
plt.title("BEFORE")
plt.subplot(1,2,2)
sns.boxplot(x_trainscale)
plt.title("AFTER")


```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
lr = LinearRegression()
lr.fit(x_trainscale,y_train)
y_pred = lr.predict(x_testscale)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)
print(f"Metrics are MAE : {mae}, MSE : {mse}, r2-score : {r2score}")
```

```python
from sklearn.linear_model import ElasticNetCV
elass = ElasticNetCV()
elass.fit(x_trainscale,y_train)
y_pred = elass.predict(x_testscale)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)
print(f"Metrics are MAE : {mae}, MSE : {mse}, r2-score : {r2score}")
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
las = Lasso()
las.fit(x_trainscale,y_train)
y_pred = las.predict(x_testscale)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)
print(f"Metrics are MAE : {mae}, MSE : {mse}, r2-score : {r2score}")

plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
rid = Ridge()
rid.fit(x_trainscale,y_train)
y_pred = rid.predict(x_testscale)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)
print(f"Metrics are MAE : {mae}, MSE : {mse}, r2-score : {r2score}")

plt.scatter(y_test,y_pred)
```
